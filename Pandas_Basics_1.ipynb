Leia novamente o arquivo passando o nome das colunas como argumento.

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
               names=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class'], header=0)
```

```python
df
```

Crie uma coluna para o volume sabendo que ele é calculado por (pi x petallength x sepal_length^2)/3

```python
df['vol']=((np.pi)*(df.iloc[:,2])*(df.iloc[:,0]**2))/3
```

```python
df
```

Calcule a média de cada atributo das flores.

```python
med=pd.DataFrame([np.mean(df.iloc[:,0]),np.mean(df.iloc[:,1]),np.mean(df.iloc[:,2]),np.mean(df.iloc[:,3]),np.mean(df.iloc[:,5])],
                 index=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','vol'],columns=['Total_Mean'])
```

```python
med
```

Novamente, calcule a média de cada atributo das flores só que agora faça-o separadamente para cada um dos tipos de íris indicados na coluna "class".

```python
df2=df.set_index('Class').copy()
setosa=[np.mean(df2.loc['Iris-setosa','Sepal_Length']),
        np.mean(df2.loc['Iris-setosa','Sepal_Width']),
        np.mean(df2.loc['Iris-setosa','Petal_Length']),
        np.mean(df2.loc['Iris-setosa','Petal_Width']),
        np.mean(df2.loc['Iris-setosa','vol'])]
versicolor=[np.mean(df2.loc['Iris-versicolor','Sepal_Length']),
            np.mean(df2.loc['Iris-versicolor','Sepal_Width']),
            np.mean(df2.loc['Iris-versicolor','Petal_Length']),
            np.mean(df2.loc['Iris-versicolor','Petal_Width']),
            np.mean(df2.loc['Iris-versicolor','vol'])]
virginica=[np.mean(df2.loc['Iris-virginica','Sepal_Length']),
           np.mean(df2.loc['Iris-virginica','Sepal_Width']),
           np.mean(df2.loc['Iris-virginica','Petal_Length']),
           np.mean(df2.loc['Iris-virginica','Petal_Width']),
           np.mean(df2.loc['Iris-virginica','vol'])]
data={'Iris-setosa_Mean':setosa,'Iris-versicolor_Mean':versicolor,'Iris-virginica_Mean':virginica}
med2=pd.DataFrame(data,
                 index=['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','vol'],
                 columns=['Iris-setosa_Mean','Iris-versicolor_Mean','Iris-virginica_Mean'])
```

```python
med2
```

Adicione uma nova coluna com o valor da média de 'sepal length' apenas para a íris da classe Iris-setosa.

```python
df['Sepal_Length_Mean']=None
for i in range (len(df)):
    if df.iloc[i,4]=='Iris-setosa':
        df.iloc[i,6]=med2.iloc[0,0]
```

```python
df
```

Adicione uma coluna com a média de 'sepal length' de tal forma que cada tipo de íris tenha seu respectivo valor de média.

```python
df['Sepal_Length_Mean']=None
for i in range (len(df)):
    if df.iloc[i,4]=='Iris-setosa':
        df.iloc[i,6]=med2.iloc[0,0]
    elif df.iloc[i,4]=='Iris-versicolor':
        df.iloc[i,6]=med2.iloc[0,1]
    elif df.iloc[i,4]=='Iris-virginica':
        df.iloc[i,6]=med2.iloc[0,2]
```
